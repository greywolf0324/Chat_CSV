```python

from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    pipeline,
    logging,
    GenerationConfig,
    TextIteratorStreamer,
)

from transformers import StoppingCriteria, StoppingCriteriaList

import torch

model_id = 'NickyNicky/gemma-1.1-2b-it_text_to_sql_format_chatML_V1'

model = AutoModelForCausalLM.from_pretrained(model_id,
                                             device_map="auto",
                                            #  trust_remote_code=True,
                                             torch_dtype=torch.bfloat16,
                                            #  attn_implementation="flash_attention_2",
                                             # load_in_4bit=True,
                                             # low_cpu_mem_usage= True,

                                             ).eval()

max_length=1355
print("max_length",max_length)


tokenizer = AutoTokenizer.from_pretrained(model_id,
                                          # use_fast = False,
                                          max_length=max_length,
                                          # trust_remote_code=True,
                                          )


class ListOfTokensStoppingCriteria(StoppingCriteria):
    """
    Clase para definir un criterio de parada basado en una lista de tokens específicos.
    """
    def __init__(self, tokenizer, stop_tokens):
        self.tokenizer = tokenizer
        # Codifica cada token de parada y guarda sus IDs en una lista
        self.stop_token_ids_list = [tokenizer.encode(stop_token, add_special_tokens=False) for stop_token in stop_tokens]

    def __call__(self, input_ids, scores, **kwargs):
        # Verifica si los últimos tokens generados coinciden con alguno de los conjuntos de tokens de parada
        for stop_token_ids in self.stop_token_ids_list:
            len_stop_tokens = len(stop_token_ids)
            if len(input_ids[0]) >= len_stop_tokens:
                if input_ids[0, -len_stop_tokens:].tolist() == stop_token_ids:
                    return True
        return False

# Uso del criterio de parada personalizado
stop_tokens = ["<end_of_turn>"]  # Lista de tokens de parada

# Inicializa tu criterio de parada con el tokenizer y la lista de tokens de parada
stopping_criteria = ListOfTokensStoppingCriteria(tokenizer, stop_tokens)

# Añade tu criterio de parada a una StoppingCriteriaList
stopping_criteria_list = StoppingCriteriaList([stopping_criteria])
```

```python
import gradio as gr
import json


import gradio as gr

import json



from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    pipeline,
    logging,
    GenerationConfig,
    TextIteratorStreamer,
)

from transformers import StoppingCriteria, StoppingCriteriaList

import torch

model_id = 'NickyNicky/gemma-1.1-2b-it_text_to_sql_format_chatML_V1'

model = AutoModelForCausalLM.from_pretrained(model_id,
                                             device_map="auto",
                                            #  trust_remote_code=True,
                                             torch_dtype=torch.bfloat16,
                                            #  attn_implementation="flash_attention_2",
                                             # load_in_4bit=True,
                                             # low_cpu_mem_usage= True,

                                             ).eval()

max_length=1355
print("max_length",max_length)


tokenizer = AutoTokenizer.from_pretrained(model_id,
                                          # use_fast = False,
                                          max_length=max_length,
                                          # trust_remote_code=True,
                                          )


class ListOfTokensStoppingCriteria(StoppingCriteria):
    """
    Clase para definir un criterio de parada basado en una lista de tokens específicos.
    """
    def __init__(self, tokenizer, stop_tokens):
        self.tokenizer = tokenizer
        # Codifica cada token de parada y guarda sus IDs en una lista
        self.stop_token_ids_list = [tokenizer.encode(stop_token, add_special_tokens=False) for stop_token in stop_tokens]

    def __call__(self, input_ids, scores, **kwargs):
        # Verifica si los últimos tokens generados coinciden con alguno de los conjuntos de tokens de parada
        for stop_token_ids in self.stop_token_ids_list:
            len_stop_tokens = len(stop_token_ids)
            if len(input_ids[0]) >= len_stop_tokens:
                if input_ids[0, -len_stop_tokens:].tolist() == stop_token_ids:
                    return True
        return False

# Uso del criterio de parada personalizado
stop_tokens = ["<end_of_turn>"]  # Lista de tokens de parada

# Inicializa tu criterio de parada con el tokenizer y la lista de tokens de parada
stopping_criteria = ListOfTokensStoppingCriteria(tokenizer, stop_tokens)

# Añade tu criterio de parada a una StoppingCriteriaList
stopping_criteria_list = StoppingCriteriaList([stopping_criteria])



def generate_text(question, context):
    question= question.replace("?","").replace("¿","").strip()
    context= context.replace('; ', ';\n').strip()

    text= f"""<bos><start_of_turn>system
You are a helpful AI assistant.
you are a sql expert who responds in json format.<end_of_turn>
<start_of_turn>user
## prompt:
{question}?

## sql context:
{context}<end_of_turn>
<start_of_turn>model
"""

    inputs = tokenizer.encode(text,
                              return_tensors="pt",
                              add_special_tokens=False).to(model.device)
    max_new_tokens=1000
    generation_config = GenerationConfig(
                  max_new_tokens=max_new_tokens,
                  temperature=0.30,
                  #top_p=0.9,
                  top_k=50,
                  repetition_penalty=1.,
                  do_sample=True,
              )
    outputs = model.generate(generation_config=generation_config,
                            input_ids=inputs,
                            stopping_criteria=stopping_criteria_list,)
    out_text= tokenizer.decode(outputs[0], skip_special_tokens=False) #True
    return out_text

def mostrar_respuesta(question, context):
    # print(question)
    # print(context)
    try:
      if question!= "" and context!= "":
        res= generate_text(question, context)
        # print(res)
        inicio_json = res.find('{')
        fin_json = res.rfind('}') + 1
        json_str = res[inicio_json:fin_json]
        json_obj = json.loads(json_str)
        # print(json_obj)
        return list(json_obj.values())
      else:
        return ["Complete los campos de pregunta y contexto correctamente" for i in range(0,9)]
    except:
      return None

context= """CREATE TABLE IF NOT EXISTS sales (equipment_type varchar(20), quantity_sold int, company varchar(30), region varchar(30));
INSERT INTO sales (equipment_type, quantity_sold, company, region) VALUES ('tank', 150, 'DefensaGlobal', 'Middle East'), ('jeep', 75, 'DefensaGlobal', 'Middle East'), ('truck', 120, 'DefensaGlobal', 'Middle East');"""

question="What is the min number of ground vehicles sold by DefensaGlobal in the Middle East?"


# Ejemplos de preguntas
ejemplos = [
    [question,context],
]

description= """Consulta sql"""

iface = gr.Interface(
    fn=mostrar_respuesta,
    inputs=[
        gr.Textbox(label="Escribe tu pregunta.", placeholder="Introduce tu pregunta en lenguaje natural aquí..."),
        gr.Textbox(label="Escribe tu contexto SQL", placeholder="Introduce contexto como tablas y valores aquí..."),

    ],
    outputs=[
        gr.Textbox(label="sql_explanation".replace("_"," "), lines=4),
        gr.Textbox(label="sql_complexity_description".replace("_"," "), lines=1),
        gr.Textbox(label="sql_complexity".replace("_"," "), lines=1),
        gr.Textbox(label="sql".replace("_"," "), lines=1),
        gr.Textbox(label="domain_description".replace("_"," "), lines=1),
        gr.Textbox(label="sql_task_type_description".replace("_"," "), lines=1),
        gr.Textbox(label="sql_task_type".replace("_"," "), lines=1),
        gr.Textbox(label="domain".replace("_"," "), lines=1),
    ],
    title="Consulta lenguaje natural SQL Ingles.",
    #description="Introduce tu diagnostico.",
    examples=ejemplos,
    concurrency_limit=20,
    description=description,
)

iface.queue(max_size=14).launch(debug=True) # share=True,debug=True
```
